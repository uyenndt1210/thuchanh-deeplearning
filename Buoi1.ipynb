## Biến một hàm Python thuần túy sang graph

```python
# Define a Python function.
def a_regular_function(x, y, b):
    x = tf.matmul(x, y)
    x = x + b
    return x

# Use tf.function to make it a TensorFlow Function
@tf.function
def a_function_that_uses_a_graph(x, y, b):
    return a_regular_function(x, y, b)

# Example usage
x1 = tf.constant([[1.0, 2.0]])
y1 = tf.constant([[2.0], [3.0]])
b1 = tf.constant(4.0)

orig_value = a_regular_function(x1, y1, b1).numpy()
tf_function_value = a_function_that_uses_a_graph(x1, y1, b1).numpy()

print(orig_value)
print(tf_function_value)
```

## Modules, Layers và Models

```python
class SimpleModule(tf.Module):
    def __init__(self, name=None):
        super().__init__(name=name)
        self.a_variable = tf.Variable(5.0, name="train_me")
        self.non_trainable_variable = tf.Variable(5.0, trainable=False, name="do_not_train_me")

    def __call__(self, x):
        return self.a_variable * x + self.non_trainable_variable

simple_module = SimpleModule(name="simple")
rs = simple_module(tf.constant(5.0))
print(rs)
print('Train variables:', simple_module.trainable_variables)
print('All variables:', simple_module.variables)
```

```python
class Dense(tf.Module):
    def __init__(self, in_features, out_features, name=None):
        super().__init__(name=name)
        self.w = tf.Variable(tf.random.normal([in_features, out_features]), name='w')
        self.b = tf.Variable(tf.zeros([out_features]), name='b')

    def __call__(self, x):
        y = tf.matmul(x, self.w) + self.b
        return tf.nn.relu(y)

class SequentialModule(tf.Module):
    def __init__(self, name=None):
        super().__init__(name=name)
        self.dense_1 = Dense(in_features=3, out_features=3)
        self.dense_2 = Dense(in_features=3, out_features=2)

    def __call__(self, x):
        x = self.dense_1(x)
        return self.dense_2(x)

my_model = SequentialModule(name="the_model")
print(my_model(tf.constant([[2.0, 2.0, 2.0]])))
```

## Lưu checkpoint

```python
chkp_path = "my_checkpoint"
checkpoint = tf.train.Checkpoint(model=my_model)
checkpoint.write(chkp_path)
print(chkp_path)
```

## Phục hồi checkpoint

```python
new_model = SequentialModule()
new_checkpoint = tf.train.Checkpoint(model=new_model)
new_checkpoint.restore("my_checkpoint")

print(new_model(tf.constant([[2.0, 2.0, 2.0]])))
```

## Bài 12

```python
import tensorflow as tf


def create_tensors():
    x = tf.random.uniform([3, 2, 2], minval=1, maxval=10, dtype=tf.float32)
    y = tf.random.uniform([3, 2, 2], minval=1, maxval=10, dtype=tf.float32)
    return x, y


@tf.function
def tensor_calculations(x, y):
    add = tf.add(x, y)          # hoặc x + y
    sub = tf.subtract(x, y)     # hoặc x - y
    mul = tf.multiply(x, y)     # hoặc x * y
    div = tf.divide(x, y)       # chia an toàn
    return add, sub, mul, div


x, y = create_tensors()

print("Tensor x shape:", x.shape)
print("Tensor y shape:", y.shape)
print("\nTensor x:\n", x.numpy())
print("Tensor y:\n", y.numpy())

add, sub, mul, div = tensor_calculations(x, y)

print("\n=== KẾT QUẢ SAU KHI CHẠY TRONG GRAPH ===")
print("Cộng   :", add.numpy())
print("Trừ    :", sub.numpy())
print("Nhân   :", mul.numpy())
print("Chia   :", div.numpy())
```
