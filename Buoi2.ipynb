## Data cleaning

```python
# =========================
# BƯỚC 4: LÀM SẠCH DỮ LIỆU TEXT
# =========================
df["Gender"] = df["Gender"].astype(str).str.lower().str.strip()
# =========================
# BƯỚC 5: CHUẨN HÓA HEIGHT
# =========================

df["Height"] = (
    df["Height"]
    .astype(str)
    .str.replace(r"[a-zA-Z\s]", "", regex=True)
)

df["Height"] = pd.to_numeric(df["Height"], errors="coerce")

df["Height"] = np.where(
    df["Height"] < 3,
    df["Height"] * 100,
    df["Height"]
)

print(df[["Name", "Height"]])


# =========================
# BƯỚC 6: CHUẨN HÓA WEIGHT
# =========================

is_lbs = df["Weight"].astype(str).str.contains("lbs", na=False)

df["Weight"] = (
    df["Weight"]
    .astype(str)
    .str.replace(r"[a-zA-Z\s]", "", regex=True)
)

df["Weight"] = pd.to_numeric(df["Weight"], errors="coerce")

df["Weight"] = np.where(
    is_lbs,
    df["Weight"] * 0.453592,
    df["Weight"]
)

print(df[["Name", "Weight"]])


# =========================
# BƯỚC 7: CHUẨN HÓA ExerciseHours
# =========================
df["ExerciseHours"] = pd.to_numeric(
    df["ExerciseHours"],
    errors="coerce"
)

print(df[["Name", "ExerciseHours"]])

df["Age"] = df["Age"].fillna(df["Age"].mean())

df["Height"] = df["Height"].fillna(df["Height"].mean())

df["Weight"] = df["Weight"].fillna(df["Weight"].mean())

df["ExerciseHours"] = df["ExerciseHours"].fillna(
    df["ExerciseHours"].mean()
)

# =========================
# BƯỚC 8: XÓA DỮ LIỆU TRÙNG
# =========================

df = df.drop_duplicates()

print(df)

encoder = LabelEncoder()

df["Gender"] = encoder.fit_transform(df["Gender"])

print(df[["Gender"]])
```

```python
import numpy as np

# ==========================================
# 1. KHỞI TẠO THAM SỐ (Weights và Bias)
# ==========================================
# X_train của bạn có 3 cột (Age, Height, Weight)
num_features = X_train.shape[1]

# Khởi tạo ngẫu nhiên hoặc bằng 0
W = np.zeros(num_features)
b = 0.0

# Thiết lập siêu tham số (Hyperparameters)
learning_rate = 0.01  # Tốc độ học
epochs = 100          # Số lượt học
m = len(X_train)      # Số lượng mẫu dữ liệu (dòng) trong tập Train

# Chuyển đổi y_train, y_test về dạng mảng NumPy để tính toán không bị lỗi
y_train_arr = np.array(y_train)
y_test_arr = np.array(y_test)

# ==========================================
# 2. VÒNG LẶP HUẤN LUYỆN (Gradient Descent)
# ==========================================
for epoch in range(epochs):
    # Bước 2.1: Dự đoán đầu ra (Forward Pass)
    y_pred_train = np.dot(X_train, W) + b

    # Bước 2.2: Tính toán Đạo hàm (Gradients) theo công thức toán
    dW = (2/m) * np.dot(X_train.T, (y_pred_train - y_train_arr))
    db = (2/m) * np.sum(y_pred_train - y_train_arr)

    # Bước 2.3: Cập nhật lại Weights và Bias (Hạ dốc)
    W = W - learning_rate * dW
    b = b - learning_rate * db

    # (Tùy chọn) In ra sai số sau mỗi 20 lượt để theo dõi
    if (epoch + 1) % 20 == 0:
        train_loss = np.mean((y_pred_train - y_train_arr) ** 2)
        print(f"Epoch {epoch+1}/{epochs} - Train Loss (MSE): {train_loss:.4f}")

print("\n--- Huấn luyện hoàn tất! ---\n")

# ==========================================
# 3. ĐÁNH GIÁ MÔ HÌNH TRÊN TẬP TEST (Evaluate)
# ==========================================
# Cho mô hình chạy trên tập X_test bằng bộ W và b đã tối ưu ở trên
y_pred_test = np.dot(X_test, W) + b

# Tự tính toán Loss (MSE) và MAE theo đúng công thức định nghĩa
loss = np.mean((y_pred_test - y_test_arr) ** 2)      # Mean Squared Error
mae = np.mean(np.abs(y_pred_test - y_test_arr))     # Mean Absolute Error

print("Loss (MSE) trên tập Test:", loss)
print("MAE trên tập Test:", mae)
```
